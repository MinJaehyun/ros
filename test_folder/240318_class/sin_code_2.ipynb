### 삼각 함수를 재사용하기 위해 함수로 작성

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
t = np.arange(0, 6, 0.01)
```

```python
t.shape
```

```python
def draw_sin(t, A, f, b):
    y = A * np.sin(np.pi * f * t) + b

    plt.figure(figsize=(12, 6))
    plt.plot(t, y)
    plt.grid()
    plt.show()
```

```python
draw_sin(t, 1, 1, 0)
```
